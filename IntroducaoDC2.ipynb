```python
import pandas as pd
```

```python
notas = pd.read_csv("tmdb_5000_movies.csv")
```

```python
notas.head()
```

```python
notas.original_language
```

```python
notas["original_language"].unique()
```

```python
notas.original_language.value_counts() #series
```

```python
notas.original_language.value_counts().index
```

```python
notas.original_language.value_counts().values
```

```python
notas.original_language.value_counts().to_frame() #data frame
```

```python
contagem_de_linguas = notas.original_language.value_counts().to_frame().reset_index() #retira o indice da primeira coluna
contagem_de_linguas.columns = ["original_language", "total"]
contagem_de_linguas.head( )
```

```python
import seaborn as sns 

```

```python
sns.barplot(x = "original_language",y = "total", data = contagem_de_linguas)
```

```python
sns.catplot(x = "original_language", kind = "count", data = notas)
```
