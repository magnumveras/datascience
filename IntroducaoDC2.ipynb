```python
contagem_de_linguas = notas.original_language.value_counts().to_frame().reset_index() #retira o indice da primeira coluna
contagem_de_linguas.columns = ["original_language", "total"]
contagem_de_linguas.head( )
```

```python
import seaborn as sns 

```

```python
sns.barplot(x = "original_language",y = "total", data = contagem_de_linguas)
```

```python
sns.catplot(x = "original_language", kind = "count", data = notas)
```

```python
print(sns.__version__)
```

```python
total_por_linguas = notas.original_language.value_counts()
total_geral = total_por_linguas.sum()
print(total_geral)

```

```python
total_ingles = total_por_linguas.loc["en"]
total_do_resto = total_geral - total_ingles
print(total_do_resto)
```

```python
dados = {
    'lingua' : ['inglês', 'outros'],
    'total' : [total_ingles, total_do_resto]
}

dados
```

```python
pd.DataFrame(dados)
```

```python
sns.barplot(x="lingua", y="total", data = dados)
```

```python
notas.query("original_language != 'en'")
```

```python
total_notas_sem_ingles = notas.query("original_language != 'en'").original_language.value_counts()
```

```python
total_por_lingua_sem_ingles = notas.query("original_language != 'en'")
total_por_lingua_sem_ingles.head()
```

```python
sns.catplot(x = "original_language", kind="count", data = total_por_lingua_sem_ingles)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(x="total", labels="lingua", data=dados)
```

```python
sns.catplot(x = "original_language", kind="count", data = total_por_lingua_sem_ingles, aspect = 2)
```

```python
sns.catplot(x = "original_language", kind="count", data = total_por_lingua_sem_ingles,
            aspect = 2,
            order = total_notas_sem_ingles.index)
```

```python
sns.catplot(x = "original_language", kind="count", data = total_por_lingua_sem_ingles,
            aspect = 2,
            palette="GnBu_d",
            order = total_notas_sem_ingles.index)
```
